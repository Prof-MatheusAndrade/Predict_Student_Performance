```python
import pandas as pd 
from docx import Document
```

```python
def switch_f92(value):
  if ((value > 0) and (value < 1)):
      return 'variou positivamente'
  elif ((value > -1) and (value < 0)):
      return 'variou negativamente'
  elif (value == 0):
      return 'permaneceu estável'
  elif (value >= 1):
      return 'apresentou alta de'
  elif (value <= -1):
      return 'apresentou baixa de'
  else:
      return 'erro'

def switch_g92(value):
  if ((value > 0) and (value < 1)):
      return 'variação positiva de'
  elif ((value > -1) and (value < 0)):
      return 'variou negativa de'
  elif (value >= 1):
      return 'avanço de'
  elif (value <= -1):
      return 'recuo de'
  else:
      return 'erro'

def switch_i92(value):
  if ((value > 0) and (value < 1)):
      return 'variação positiva de'
  elif ((value > -1) and (value < 0)):
      return 'variou negativa de'
  elif (value >= 1):
      return 'alta de'
  elif (value <= -1):
      return '‘baixa de'
  else:
      return 'erro'

def switch_l92(value):
  if ((value > 0) and (value < 1)):
      return 'apresentou incremento de'
  elif ((value > -1) and (value < 0)):
      return 'apresentou variação negativa de'
  elif (value == 0):
      return 'permaneceu estável'
  elif (value >= 1):
      return 'apresentou alta de'
  elif (value <= -1):
      return '‘apresentou queda de'
  else:
      return 'erro'

def switch_m92(value):
  if ((value > 0) and (value < 1)):
      return 'variou positivamente'
  elif ((value > -1) and (value < 0)):
      return 'variou negativamente'
  elif (value >= 1):
      return 'avançou'
  elif (value <= -1):
      return 'retrocedeu'
  else:
      return 'erro'

def switch_n92(value):
  if ((value > 0) and (value < 1)):
      return 'variação positiva de'
  elif ((value > -1) and (value < 0)):
      return 'variação negativa de'
  elif (value >= 1):
      return 'avanço de'
  elif (value <= -1):
      return 'recuo de'
  else:
      return 'erro'

def switch_m200_n200(value):
  if ((value > 0) and (value < 1)):
      return 'variação positiva de'
  elif ((value > -1) and (value < 0)):
      return 'variação negativa de'
  elif (value >= 1):
      return 'acréscimo de'
  elif (value <= -1):
      return 'decréscimo de'
  else:
      return 'erro'
  
def d96_cond_semana(value_itaqui, value_aratu, value_paranagua, value_suape, value_santos, value_tramandai):
  if((value_itaqui<0) & (value_aratu<0) & (value_paranagua<0) & (value_suape<0) & (value_santos<0) & (value_tramandai<0)):
    return 'negativas'
  elif((value_itaqui>0) & (value_aratu>0) & (value_paranagua>0) & (value_suape>0) & (value_santos>0) & (value_tramandai>0)):
    return 'positivas'
  else:
    return ''

def d96_cond_menor(value_itaqui, value_aratu, value_paranagua, value_suape, value_santos, value_tramandai):
  list_value = [value_itaqui, value_aratu, value_paranagua, value_suape, value_santos, value_tramandai]
  df_aux = pd.DataFrame({'value': list_value},
                   index=['Itaqui', 'Aratu', 'Paranagua','Suape','Santos','Tramandaí'])
  return min(list_value),df_aux.idxmin()['value']

def d96_cond_maior(value_itaqui, value_aratu, value_paranagua, value_suape, value_santos, value_tramandai):
  list_value = [value_itaqui, value_aratu, value_paranagua, value_suape, value_santos, value_tramandai]
  df_aux = pd.DataFrame({'value': list_value},
                   index=['Itaqui', 'Aratu', 'Paranagua','Suape','Santos','Tramandaí'])
  return max(list_value),df_aux.idxmax()['value']
```

```python
df = pd.read_excel('Relatorio-Sintese-Ordinaria nº 07_KPO.xlsx')

e85_i85 = df.iloc[83,4]

f92 = abs(df.iloc[90,5] * 100) 
f92 = "{:.2f}".format(f92).replace('.',',') + '%' + ' '
f92_cond = switch_f92((df.iloc[90,5] * 100)) + ' '

e92 = abs(df.iloc[90,4])
e92 = "{:.2f}".format(e92).replace('.',',')

g92_cond = switch_g92((df.iloc[90,6] * 100)) + ' '
g92 = abs(df.iloc[90,6] * 100)
g92 = "{:.2f}".format(g92).replace('.',',') + '%' + ' '

i92_cond = switch_i92((df.iloc[90,8] * 100)) + ' '
i92 = abs(df.iloc[90,8] * 100)
i92 = "{:.2f}".format(i92).replace('.',',') + '%' + ' '

k85_n85 = df.iloc[83,10]

l92_cond = switch_l92((df.iloc[90,11] * 100)) + ' '
l92 = abs(df.iloc[90,11] * 100) 
l92 = "{:.2f}".format(l92).replace('.',',') + '%' + ' '

k92 = abs(df.iloc[90,8])
k92 = "{:.2f}".format(k92).replace('.',',')

m92_cond = switch_m92((df.iloc[90,12] * 100)) + ' '
m92 = abs(df.iloc[90,12] * 100)
m92 = "{:.2f}".format(m92).replace('.',',') + '%' + ' '

n92_cond = switch_n92((df.iloc[90,13] * 100)) + ' '
n92 = abs(df.iloc[90,13] * 100)
n92 = "{:.2f}".format(n92).replace('.',',') + '%' + ' '

i200_l200 = abs(df.iloc[198,8])
i200_l200 = "{:.2f}".format(i200_l200).replace('.',',')

m200_n200_cond = switch_m200_n200((df.iloc[90,12] * 100)) + ' '
m200_n200 = abs(df.iloc[198,12])
m200_n200 = "{:.2f}".format(m200_n200).replace('.',',') + '%' + ' '

d96_cond = d96_cond_semana((df.iloc[94,3] * 100), (df.iloc[94,8] * 100), (df.iloc[94,13] * 100), 
                           (df.iloc[95,3] * 100), (df.iloc[95,8] * 100), (df.iloc[95,13] * 100))

d96_menor, d96_menor_porto = d96_cond_menor((df.iloc[94,3] * 100), (df.iloc[94,8] * 100), (df.iloc[94,13] * 100), 
                           (df.iloc[95,3] * 100), (df.iloc[95,8] * 100), (df.iloc[95,13] * 100))

d96_maior, d96_maior_porto = d96_cond_maior((df.iloc[94,3] * 100), (df.iloc[94,8] * 100), (df.iloc[94,13] * 100), 
                           (df.iloc[95,3] * 100), (df.iloc[95,8] * 100), (df.iloc[95,13] * 100))

d96_menor = abs(d96_menor)
d96_menor = "{:.2f}".format(d96_menor).replace('.',',') + '%' + ' '

d96_menor_porto = str(d96_menor_porto)

d96_maior = abs(d96_maior)
d96_maior = "{:.2f}".format(d96_maior).replace('.',',') + '%' + ' '

d96_maior_porto = str(d96_maior_porto)
```
